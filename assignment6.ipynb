```python
import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
cds=pd.read_stata('C:\\Users\\Tina Sun\\Desktop\\term3\\machine learning\\machine-learning\\cds_spread5y_2001_2016.dta')
qtr=pd.read_csv('C:\\Users\\Tina Sun\\Desktop\\term3\\machine learning\\machine-learning\\qtr_crsp2.csv')
```

```python
#make sure the gvkey of this two file are same
cds.gvkey.dtypes
qtr["GVKEY"].dtypes
qtr.columns = qtr.columns.str.lower()
cds.gvkey=cds.gvkey.astype('int64')
cds.gvkey.dtypes
cds.shape
cds.columns.values
```

```python
#generate a new list of quarters for cds 
cds.mdate.dtypes  #<M8[ns]
qtr.datadate.dtypes   #int64
cds_qtr=list()
i=0 
cds.size
c=cds['mdate']
c.size

m=cds['mdate'].dt.month
y=cds['mdate'].dt.year
for i in range(0,int(c.size)):
    mm=m[i]
    yy=y[i]
    if mm>0 and mm<4:
        a=str(yy)+'Q1'
    if mm>3 and mm<7:
        a=str(yy)+'Q2'
    if mm>6 and mm<10:
        a=str(yy)+'Q3'
    else:
        a=str(yy)+'Q4'     
    cds_qtr.append(a)
    i=i+1
cds_qtr 
```

```python
cds_qtr=pd.Series(cds_qtr)
cds1 = cds.assign(datacqtr=cds_qtr.values) 
cds1
```

```python
#merge two dataframe
new=pd.merge(cds1, qtr, on=['gvkey', 'datacqtr'])
new
```

```python
#data cleaning
new1=new.dropna(how='all')
new1=new1._get_numeric_data()
new1=new1.fillna(new1.median(),inplace=True)
frac=len(new1)*0.5
new1=new1.dropna(thresh=frac,axis=1)
new1
```
