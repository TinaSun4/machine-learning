```python
import pandas as pd
qtr=pd.read_csv('C:\\Users\\Tina Sun\\Desktop\\term3\\machine learning\\machine-learning\\qtr_crsp2.csv')
cds=pd.read_stata('C:\\Users\\Tina Sun\\Desktop\\term3\\machine learning\\machine-learning\\cds_spread5y_2001_2016.dta')
```

```python
#make sure the gvkey of this two file are same
cds.gvkey.dtypes
qtr["GVKEY"].dtypes
qtr.columns = qtr.columns.str.lower()
cds.gvkey=cds.gvkey.astype('int64')
cds.gvkey.dtypes
cds
```

```python
#generate a new list of quarters for cds 
cds.mdate.dtypes  #<M8[ns]
qtr.datadate.dtypes   #int64
cds_qtr=list()
i=0 
cds.size
c=cds['mdate']
c.size

m=cds['mdate'].dt.month
y=cds['mdate'].dt.year
for i in range(0,int(c.size)):
    mm=m[i]
    yy=y[i]
    if mm>0 and mm<4:
        a=str(yy)+'Q1'
    if mm>3 and mm<7:
        a=str(yy)+'Q2'
    if mm>6 and mm<10:
        a=str(yy)+'Q3'
    else:
        a=str(yy)+'Q4'     
    cds_qtr.append(a)
    i=i+1
cds_qtr 
```

```python
cds_qtr=pd.Series(cds_qtr)
cds1 = cds.assign(datacqtr=cds_qtr.values) 
cds1
```

```python
#merge two dataframe
new=pd.merge(cds1, qtr, on=['gvkey', 'datacqtr'])
new
```

```python
#data cleaning
new1=new.dropna(how='all')
new1=new1._get_numeric_data()
new1=new1.fillna(new1.median(),inplace=True)
frac=len(new1)*0.5
new1=new1.dropna(thresh=frac,axis=1)
new1
```

```python
#Divide the dataset to test and train. Consider 2016, 2017, and 2018 data as test and the rest as train.
#Y_column = 'spread5y'
test = new1[new1['fyearq']>=2016]
train = new1[new1['fyearq']<2016]
train.shape  # (105163, 641)
test.shape  #(7213, 641)
```

```python
y_train=train['spread5y']
x_train=train.drop('spread5y',axis=1)
y_test=test['spread5y']
x_test=test.drop('spread5y',axis=1)
```

```python
## On the train sample run a Random Forest with 50 trees
```

```python
from sklearn import ensemble
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.datasets import make_regression
```

```python
x_test.shape
```

```python
pd.set_option('display.max_columns', 100)
x_train=pd.DataFrame(data=x_train)
missing_columns = x_train.loc[:,x_train.iloc[1,:].isnull()].columns.values
x_train_miss = x_train.drop(missing_columns,axis=1)
x_test_miss = x_test.drop(missing_columns,axis=1)
x_train_miss.shape (105163, 640)
```

```python
pd.set_option('display.max_columns', 1000)
x_train=pd.DataFrame(data=x_train)
missing_columns = x_train.loc[:,x_train.iloc[1,:].isnull()].columns.values
x_train_miss = x_train.drop(missing_columns,axis=1)
x_test_miss = x_test.drop(missing_columns,axis=1)
x_train_miss.shape
```

```python
x_train_miss, y_train = make_regression(n_features=640, n_informative=2,random_state=0, shuffle=False)
regr = RandomForestRegressor(max_depth=3, random_state=0,n_estimators=50)
regr.fit(x_train_miss, y_train)
```

```python
print(regr.feature_importances_)
type(regr.feature_importances_)
regr.feature_importances_.size  # 640
```

```python
#keep the first 50 most important feature in train dataset
imp=list(regr.feature_importances_)
x_train1=pd.DataFrame(data=x_train_miss).T
imp=pd.Series(imp)
fea=x_train1.assign(importance=imp.values) #add one more columns about feature importance
f_50=fea.sort_values('importance',ascending=False)[0:50]
f_50=f_50.drop('importance',axis=1) #50 rows of features
train_50=f_50.T
train_50
```

```python
x_train1.shape #640 rows
```

```python
#keep the first 50 variable in test dataset
imp=list(regr.feature_importances_)
x_test1=pd.DataFrame(data=x_test_miss).T
imp=pd.Series(imp)
fea1=x_test1.assign(importance=imp.values) #620 rows
test_50=fea1.sort_values('importance',ascending=False)[0:50]
test_50=test_50.drop('importance',axis=1)#50 rows of features
test_50=test_50.T 
test_50
```

```python

#Using these 50 variables run Random Forest, Gradient Boosting, and XGBoost with 100, 200, 500, and 1000 trees.
##Random Forest
from sklearn import ensemble
from sklearn.metrics import mean_squared_error
y=train['spread5y']
x=train_50
x, y = make_regression(n_features=50, n_informative=2,random_state=0, shuffle=False)
#100 trees
rf100=RandomForestRegressor(max_depth=3, random_state=0,n_estimators=100)
rf100.fit(x, y)
mse_rf100 = mean_squared_error(y_test, rf100.predict(test_50))
print("Random tree with 100 trees:MSE: %.4f" % mse_rf100)

#rf200=RandomForestRegressor(max_depth=3, random_state=0,n_estimators=200)
rf200=RandomForestRegressor(max_depth=3, random_state=0,n_estimators=200)
rf200.fit(x, y)
mse_rf200 = mean_squared_error(y_test, rf200.predict(test_50))
print("Random tree with 100 trees:MSE: %.4f" % mse_rf200)

#rf500=RandomForestRegressor(max_depth=3, random_state=0,n_estimators=500)
rf500=RandomForestRegressor(max_depth=3, random_state=0,n_estimators=500)
rf500.fit(x, y)
mse_rf500 = mean_squared_error(y_test, rf500.predict(test_50))
print("Random tree with 100 trees:MSE: %.4f" % mse_rf500)

#rf1000=RandomForestRegressor(max_depth=3, random_state=0,n_estimators=1000)
rf1000=RandomForestRegressor(max_depth=3, random_state=0,n_estimators=1000)
rf1000.fit(x, y)
mse_rf1000 = mean_squared_error(y_test, rf1000.predict(test_50))
print("Random tree with 100 trees:MSE: %.4f" % mse_rf1000)
```

```python
# the relationship between number of trees and â€œMean Squared Errorâ€ of test sample in a line graph in random tree
import matplotlib.pyplot as plt
from matplotlib.pyplot import plot
n=[100,200,500,1000]
mse_rf=[mse_rf100,mse_rf200,mse_rf500,mse_rf1000]
plot(n,mse_rf)
plt.show()
```

```python
##Gradient Boosting
#Gradient Boosting with 100 trees
gb100=ensemble.GradientBoostingRegressor(max_depth=3, random_state=0,n_estimators=100)
gb100.fit(x, y)
mse_gb100 = mean_squared_error(y_test, gb100.predict(test_50))
print("Gradient Boosting with 100 trees:MSE: %.4f" % mse_gb100)

#Gradient Boosting with 200 trees
gb200=ensemble.GradientBoostingRegressor(max_depth=3, random_state=0,n_estimators=200)
gb200.fit(x, y)
mse_gb200 = mean_squared_error(y_test, gb200.predict(test_50))
print("Gradient Boosting with 200 trees:MSE: %.4f" % mse_gb200)

#Gradient Boosting with 500 trees
gb500=ensemble.GradientBoostingRegressor(max_depth=3, random_state=0,n_estimators=500)
gb500.fit(x, y)
mse_gb500 = mean_squared_error(y_test, gb500.predict(test_50))
print("Gradient Boosting with 500 trees:MSE: %.4f" % mse_gb500)

#Gradient Boosting with 1000 trees
gb1000=ensemble.GradientBoostingRegressor(max_depth=3, random_state=0,n_estimators=1000)
gb1000.fit(x, y)
mse_gb1000 = mean_squared_error(y_test, gb1000.predict(test_50))
print("Gradient Boosting with 1000 trees:MSE: %.4f" % mse_gb1000)
```

```python
# the relationship between number of trees and â€œMean Squared Errorâ€ of test sample in a line graph in gradient boosting
import matplotlib.pyplot as plt
from matplotlib.pyplot import plot
n=[100,200,500,1000]
mse_gb=[mse_gb100, mse_gb200,mse_gb500,mse_gb1000]
plot(n,mse_gb)
plt.show()
```
